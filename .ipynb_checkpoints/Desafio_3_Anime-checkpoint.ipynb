```python
#Defino las columnas con las que voy a trabajar
data_util = data[['Name','English name',"Type", 'Episodes', 'Score','Popularity', 'Watching', 'Completed', 'On-Hold', 'Dropped', 'Plan to Watch']]
```

```python
#Ordeno en funcion de los valores de Watching
data_util.sort_values('Watching', ascending = False)
```

```python
#Seteo como indice el nombre del Anime
data_util.set_index(['Name'], append=True)
```

```python
#Veo cuantos son los animes que nadie esta viendo en este momento
data_util.loc[ data_util.Watching == 0 ]
```

```python
#continuo el analisis del Data set
data_util.describe()
```

```python
#Renombro la columna de Plan to Watch para poder aplicar mascaras
data_util.rename(columns={ 'Plan to Watch' : 'Plan_to_Watch' }, inplace = True)
#preguntar por que se rompe
```

```python
data_util[['Score']] = data_util[['Score']].replace({'Unknown': 0 })
```

```python
data_util[['Score']] = data_util[['Score']].astype(float).round(0)
#Finalizo la limpieza de Score para poder tener solo valores numericos y no nulos
```

```python
#Ecncuentro los animes con calificacion ya que las que presentan 0 no las conozco
data_util[['Score']].value_counts()
data_util = data_util.loc[data_util.Score >= 1 ]
data_util
```

```python
#Genero un plot de correlacion para entender cuales son las realciones entre las variables
data_limpia = data_util.drop(['Name', 'Type', 'English name', 'Episodes'], axis = 1)
data_limpia.corr()
sns.heatmap(data_limpia.corr(), cmap="BuPu", annot=True, vmin=0)
```

```python
data_util
# Creamos dummies para Score
dummies_Score = pd.get_dummies(data_util['Score'], drop_first=True, prefix='Score')
dummies_Score.head()
```

```python
data_util = pd.concat([data_util, dummies_Score], axis=1)
data_util.head()
```

```python
data_util['Anime_Bueno'] = data_util.Score[data_util.Score >= 6.0]

data_util['Anime_Malo'] = data_util.Score[data_util.Score < 6.0]
```

```python
data_util['Anime_Bueno'].fillna(0, inplace = True)
```

```python
data_util['Anime_Malo'].fillna(0, inplace = True)
```

```python
data_util['Anime_Bueno'].replace([9.0, 8.0, 7.0, 6.0], 1, inplace= True)
```

```python
data_util['Anime_Malo'].replace([5.0, 4.0, 3.0, 2.0], 1, inplace= True)
```

```python
data_util['Anime_Bueno'] = data_util['Anime_Bueno'].astype(int)
data_util['Anime_Malo'] = data_util['Anime_Malo'].astype(int)
```
